## ETAPE 3: Text Preprocessing

*** création d'une fonction qui regroupera les étapes clés du traitement du texte. Cette fonction préparera les deux jeux de données de manière cohérente, rendant le texte prêt pour les modèles de machine learning et deep learning.***

```python
import pandas as pd
import re
import string
from nltk.corpus import stopwords
import contractions

# Fonction de prétraitement globale
def preprocess_text(df, column_name):
    STOPWORDS = set(stopwords.words('english'))

    # Fonction pour étendre les contractions
    def expand_contractions(text):
        if isinstance(text, str):
            return contractions.fix(text)
        return str(text)  # Convertit en chaîne si nécessaire

    # Fonction pour retirer les chiffres
    def remove_numbers(text):
        if isinstance(text, str):
            return ''.join([i for i in text if not i.isdigit()])
        return text

    # Fonction pour convertir en minuscule
    def to_lowercase(text):
        if isinstance(text, str):
            return text.lower()
        return str(text).lower()

    # Fonction pour supprimer les stopwords
    def remove_stopwords(text):
        if isinstance(text, str):
            return " ".join([word for word in text.split() if word not in STOPWORDS])
        return text

    # Fonction pour supprimer les balises HTML
    def remove_html(text):
        if isinstance(text, str):
            html_pattern = re.compile('<.*?>')
            return html_pattern.sub(' ', text)
        return text

    # Fonction pour supprimer les ponctuations
    def remove_punctuation(text):
        if isinstance(text, str):
            return text.translate(str.maketrans('', '', string.punctuation))
        return text

    # Fonction pour supprimer les caractères spéciaux
    def remove_special_characters(text):
        if isinstance(text, str):
            return re.sub(r'[^a-zA-Z0-9\s]', '', text)
        return text

    # Fonction pour supprimer les espaces inutiles
    def remove_extra_spaces(text):
        if isinstance(text, str):
            return re.sub(r'\s+', ' ', text).strip()
        return text

    # Nettoyage initial : convertir toutes les valeurs en chaînes
    df[column_name] = df[column_name].astype(str)

    # Application séquentielle des fonctions de prétraitement
    df[column_name] = df[column_name].apply(expand_contractions)
    df[column_name] = df[column_name].apply(remove_numbers)  # Retirer les chiffres en deuxième position
    df[column_name] = df[column_name].apply(to_lowercase)
    df[column_name] = df[column_name].apply(remove_html)
    df[column_name] = df[column_name].apply(remove_punctuation)
    df[column_name] = df[column_name].apply(remove_special_characters)
    df[column_name] = df[column_name].apply(remove_stopwords)
    df[column_name] = df[column_name].apply(remove_extra_spaces)

    return df


df = preprocess_text(df, 'statement') # Appeller la fonction
```

## 1.  Analyse des Sentiments avec TextBlob

Dans cette section, nous allons utiliser **TextBlob** pour analyser chaque phrase de notre ensemble de données. TextBlob est un outil puissant de traitement du langage naturel qui calcule la **polarité** des phrases. La polarité se décompose en deux éléments principaux : **polarité** et **subjectivité**.

### 1. Polarité
La **polarité** est une valeur numérique qui identifie si une phrase exprime un sentiment **positif** ou **négatif**. Elle varie entre **-1 et 1**, où :

- **1** indique un **sentiment très positif** ou une affirmation basée sur des **faits** plutôt que des émotions.  
- **-1** reflète un **sentiment très négatif** ou une affirmation sans fondement factuel.

En résumé :  
- Plus la polarité est proche de **1**, plus le sentiment est positif.  
- Plus elle est proche de **-1**, plus le sentiment est négatif.

### 2. Subjectivité
La **subjectivité** mesure dans quelle mesure une phrase reflète des **opinions personnelles** ou des **émotions**, par opposition à des **faits objectifs**. Ses valeurs oscillent entre **0 et 1** :

- **1** indique une forte subjectivité, c’est-à-dire une phrase fondée sur des **opinions** ou des **émotions**.  
- **0** suggère une forte objectivité, indiquant une phrase basée sur des **faits**.

### Résumé
La polarité évalue le **ton émotionnel** (positif ou négatif) d’une phrase, tandis que la subjectivité détermine si l’énoncé repose sur des **opinions** ou des **faits**. Ces deux métriques combinées nous permettent d’obtenir une analyse plus riche des sentiments véhiculés dans les textes.

### En conclusion  

Rappelons que l'analyse des sentiments avec Texblob vise à différencier les messages factuels des messages émotionnels.  

- Les **messages factuels** sont généralement représentés par des **valeurs positives**.  
  - Ils reflètent des faits réels, souvent soutenus par des preuves concrètes.  
  - Exemple : Une personne anxieuse dit qu'elle est anxieuse parce qu'elle s'appuie sur des événements ou des circonstances tangibles qui justifient son ressenti.  

- En revanche, les **messages émotionnels** sont souvent représentés par des **valeurs négatives**.  
  - Ces messages expriment un ressenti dicté principalement par les émotions, sans preuves tangibles.  
  - Exemple : Une personne peut déclarer être en dépression non pas parce qu'elle l'est réellement, mais par empathie ou émotion, s'identifiant à la souffrance d'autrui.  

Cependant, cette analyse ne garantit pas une vérité absolue. Elle offre plutôt une méthode pour évaluer, en moyenne, la véracité ou l'origine des messages, tout en tenant compte de la nuance entre émotion et réalité.

```python
# Téléchargement  du  model pré-entrainé  Glove

# Chemin vers les embeddings GloVe
glove_file = "C:/Users/claud/Documents/Kaggle Data/glove.6B/glove.6B.100d.txt"

# Chargement des embeddings GloVe
embeddings_index = {}
f = open(glove_file, encoding="utf8")  
for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embeddings_index[word] = coefs
f.close()

print('Found %s word vectors.' % len(embeddings_index))
```

```python
# initialisation de la  matrice d'embedding pour le modèle 

embedding_dim = 100
embedding_matrix = np.zeros((max_words, embedding_dim))
for word, i in word_index.items():
    embedding_vector = embeddings_index.get(word)
    if i < max_words:
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector
```

```python
# Creation du Model de Deep Learning
from tensorflow.keras.layers import Input

model = Sequential()
model.add(Input(shape=(maxlen,))) 
model.add(Embedding(max_words, embedding_dim, input_length = maxlen))
model.add(Flatten())
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.1))
model.add(Dense(32, activation='sigmoid'))
model.add(Dense(7, activation='softmax'))
model.summary()
```

```python
model.layers[0].set_weights([embedding_matrix]) # Initialisation des poids de la couche d'embedding :
model.layers[0].trainable = False # Empêcher l’entraînement de la couche d’embedding :
```

```python
#Compiler le modele

model.compile(
    loss='categorical_crossentropy',  # 'categorical_crossentropy' pour une classification multi-classes
    optimizer = 'Adam', 
    metrics=['accuracy']  # Mesure la précision pendant l'entraînement
)
```

```python
# Convertir les labels en format one-hot

from keras.utils import to_categorical
y_train_one_hot = to_categorical(y_train, num_classes=7)  # 7 classes
y_val_one_hot = to_categorical(y_val, num_classes=7)  # 7 classes
```

```python
# Entrainement

history = model.fit(
    X_train,  # Données d'entraînement
    y_train_one_hot,  # Labels d'entraînement en format one-hot
    epochs=10,  # Nombre d'époques
    batch_size=32,  # Taille du batch
    validation_data=(X_val, y_val_one_hot)
  
)
```

```python
# Afficher les métrics

import matplotlib.pyplot as plt

# Affichage de la précision
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Précision du modèle')
plt.xlabel('Époques')
plt.ylabel('Précision')
plt.legend(['Entraînement', 'Validation'])
plt.show()

# Affichage de la perte
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Perte du modèle')
plt.xlabel('Époques')
plt.ylabel('Perte')
plt.legend(['Entraînement', 'Validation'])
plt.show()
```

### Performance du modèle

- **Précision d'entraînement** : 100% à partir de la 6e époque.
- **Perte d'entraînement** : Réduction constante jusqu'à 0.13.
- **Précision de validation** : Fluctue entre 49% et 53%.
- **Perte de validation** : Reste élevée, atteignant des valeurs supérieures à 1.5.

Ces résultats sont un signe flagrant de **overfitting**, où le modèle se surajuste aux données d'entraînement sans 
réussir à bien généraliser sur les données de validation.

Les résultats obtenus sont limités par plusieurs facteurs. Tout d'abord, le manque de données avec seulement 200 
échantillons pour l'entraînement empêche le modèle de capter des schémas complexes. Le déséquilibre entre entraînement 
et validation (200 échantillons d'entraînement contre 10 000 pour la validation) cause un biais dans l'évaluation. Enfin,
même avec l'utilisation de vecteurs de mots pré-entraînés comme GloVe, le modèle ne peut pas exploiter pleinement 
ces embeddings en raison du manque de données suffisantes. 
Pour améliorer les performances, une augmentation du jeu de données et un meilleur équilibre entre les ensembles

##### Pourquoi stratifier ?
La stratification est essentielle lorsque vous travaillez avec des classes déséquilibrées (par exemple, si certaines catégories comme 
***Dépression*** ou ***Stress*** apparaissent beaucoup plus fréquemment que d'autres). Cela garantit que chaque ensemble de données (entraînement et test) contient une distribution proportionnelle des classes.

```python
# Fonction d'évaluation du modéle

def metrics_score(actual, predicted, classes):
    # Générer le rapport de classification sous forme de DataFrame
    report = classification_report(actual, predicted, target_names=classes, output_dict=True)
    report_df = pd.DataFrame(report).transpose()
    
    return report_df

```

```python
# Fonction pour calculer les poids

def compute_weight(df, target):
    values = target.unique()  # Obtenir les valeurs uniques dans target (les classes)
    
    for value in values:
        n_classes = target.nunique()  # Nombre de classes uniques
        n_samples = df.shape[0]  # Nombre total d'échantillons
        class_samples = target[target == value].shape[0]  # Nombre d'échantillons pour la classe 'value'

        # Calcul du poids pour chaque classe
        class_weight = round(n_samples / (n_classes * class_samples), 2)
        
        print(f"Poids pour la classe '{value}': {class_weight}")
        
compute_weight(df, df['status'])
```

Nous avons constaté que la distribution des classes est déséquilibrée dans notre jeu de données. Dès le départ, afin d'améliorer les performances de notre modèle, nous avons décidé d'attribuer des poids spécifiques à chaque classe.

Ces poids sont calculés à l'aide de la formule suivante :

\[
w_j = \frac{n_{\text{samples}}}{n_{\text{classes}} \times n_{\text{samples}_j}}
\]

### Où :
- \(w_j\) : Poids attribué à la classe \(j\).
- \(n_{\text{samples}}\) : Nombre total d'échantillons (lignes) dans le jeu de données.
- \(n_{\text{classes}}\) : Nombre total de classes uniques dans la variable cible.
- \(n_{\text{samples}_j}\) : Nombre d'échantillons appartenant à la classe \(j\).

> **Remarque** : Ce résultat peut également être obtenu automatiquement en définissant le paramètre 
`class_weight='balanced'` dans de nombreux algorithmes d'apprentissage supervisé.


```python
class_weights = {
    'Anxiety': 1.96,
    'Normal': 0.46,
    'Depression': 0.49,
    'Suicidal': 0.71,
    'Stress': 2.91,
    'Bipolar': 2.71,
    'Personality disorder': 6.99
}
rf_estimator = RandomForestClassifier(class_weight=class_weights, random_state=1)
rf_estimator.fit(X_train, y_train)
```

```python
# Performance sur le jeu de donnée d'entrainement. 

# Liste des classes
classes = ['Normal', 'Depression', 'Suicidal', 'Anxiety', 'Bipolar', 'Stress', 'Personality disorder']

# Prédictions sur les données d'entraînement
y_pred_train_rf = rf_estimator.predict(X_train)

# Appeler la fonction avec les classes
metrics_score(y_train, y_pred_train_rf, classes)
```

```python
#  performance sur le jeu de donnée de validation

# Liste des classes
classes = ['Normal', 'Depression', 'Suicidal', 'Anxiety', 'Bipolar', 'Stress', 'Personality disorder']

# Prédictions sur les données d'evaluation
y_pred_test_rf = rf_estimator.predict(X_test)
 
# Appeler la fonction avec les classes
metrics_score(y_test, y_pred_test_rf, classes)
```

```python

# Définir les hyperparamètres à optimiser

parameters = {
    'n_estimators': [50],  # Nombre d'arbres
    'max_depth': [None, 10, 20], # Profondeur des arbres
    'min_samples_split': [2, 5], # Critère de division
    'min_samples_leaf': [1, 2]  # Nombre minimum d'échantillons dans une feuille
}

# Appliquer GridSearchCV
grid_search = GridSearchCV(estimator=rf_estimator, param_grid=parameters, cv=2, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

# Résultats de la recherche
print("Meilleurs hyperparamètres :", grid_search.best_params_)

# Prédictions avec le meilleur modèle
y_pred = grid_search.best_estimator_.predict(X_test)

# Rapport de classification
print(classification_report(y_test, y_pred))
```
